# SoccerData: Understat — Big 5, all players

This notebook uses [`soccerdata`](https://soccerdata.readthedocs.io/en/) **only** with [**Understat**](https://understat.com/) to load **player–season** stats for **all players** in the **Big 5** leagues, for the **last five seasons** (dynamically from today’s date).

European leagues use **multi-year** season codes internally (e.g. `2425` for 2024–25). Data is cached under `~/soccerdata/data/Understat`.

> First run can take a while while each league–season is downloaded; later runs use the cache.

```python
# If needed, uncomment:
# %pip install soccerdata pandas

from datetime import datetime, timezone

import pandas as pd
import soccerdata as sd

# Big 5 — soccerdata internal league ids (see soccerdata LEAGUE_DICT)
BIG5_LEAGUES = [
    "ENG-Premier League",
    "ESP-La Liga",
    "GER-Bundesliga",
    "ITA-Serie A",
    "FRA-Ligue 1",
]





SEASONS = [
           1516,
           1617,
           1718,
           1819,
           1920,
           2021,
           2122,
           2223,
           2324, 
           2425,
           2526
           ]

pd.set_option("display.max_columns", 80)
print(f"Leagues: {BIG5_LEAGUES}")
print(f"Season start years (passed to Understat): {SEASONS}")
```

```python
# Understat: player-season stats for Big 5 × last 5 seasons
# Docs: https://soccerdata.readthedocs.io/en/latest/reference/understat.html
understat = sd.Understat(leagues=BIG5_LEAGUES, seasons=SEASONS)

understat_player_season = understat.read_player_season_stats()

print("Understat player-season shape:", understat_player_season.shape)
understat_player_season.head(10)
```

```python
# Quick sanity check: row counts by league and season
counts = (
    understat_player_season.reset_index()
    .groupby(["league", "season"], observed=True)
    .size()
    .rename("n_players")
    .reset_index()
    .sort_values(["league", "season"])
)
counts
```

```python
understat_player_season.to_csv("./data/understat_player_season.csv")
```
